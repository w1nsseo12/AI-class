```python
# 2x4 subplot ì´ìš©. axsëŠ” 4x2
fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)

lm_features = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']

# iì—ëŠ” ì¸ë±ìŠ¤ê°€ featureì—ëŠ” RM ~ RADê¹Œì§€ ìˆœì°¨ì ìœ¼ë¡œ ë“¤ì–´ê°
for i, feature in enumerate(lm_features):
    row = int(i/4) # 2í–‰
    col = i%4

    # sns.regplot : íšŒê·€ì§ì„ ì„ ê·¸ë ¤ì¤Œ
    sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_data=df.loc[:, ['LSTAT', 'RM']]
y_data=df.loc[:, 'PRICE']
X_train, X_test, y_train, y_test=train_test_split(x_data,
                                                  y_data,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#StandardScaler()

from sklearn.preprocessing import StandardScaler

std = StandardScaler()
std.fit(X_train)
X_train_scaled = std.transform(X_train)
X_test_scaled = std.transform(X_test)
std.fit(X_train_scaled, y_train)
```

```python
# í›ˆë ¨ ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ë¥¼ 2ì°¨ì› ë°°ì—´ë¡œ ë°”ê¿‰ë‹ˆë‹¤
X_train = X_train.reshape(-1, 1)
X_test =  X_test.reshape(-1, 1)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=3)

knr.fit(X_train, y_train)
```

```python
#ì˜ˆì¸¡ê°’ ì €ìž¥
y_pred = regressor.predict(X_test) 
y_pred
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
print("íšŒê·€ê³„ìˆ˜(ê¸°ìš¸ê¸°):", np.round(lr.coef_, 1)) #np: numpy, coef_: í”¼ì²˜ì— ëŒ€í•œ íšŒê·€ ê³„ìˆ˜ ê°’
print("ìƒìˆ˜í•­(ì ˆíŽ¸):", np.round(lr.intercept_, 1))  #intercept_: ìƒìˆ˜í•­(ì ˆíŽ¸) ê°’
```

```python
y_preds = lr.predict(X_test)
print(X_train.shape, X_test.shape)
```

```python
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
print(f'MSE : {mse:.3f}, RMSE: {rmse:.3f}')
print(f'Variance score : {r2_score(y_test, y_preds):.3f}')
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X_test['LSTAT'], y_test, label='y_test')
plt.scatter(X_test['LSTAT'], y_preds, c='r', label='y_pred')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()   
```

```python
print("ì ˆíŽ¸ ê°’:", lr.intercept_) # yì¶• ì ˆíŽ¸ê°’

# íšŒê·€ ê³„ìˆ˜(coefficient) : ë…ë¦½ë³€ìˆ˜ì˜ ë³€í™”ì— ë”°ë¼ ì¢…ì†ë³€ìˆ˜ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ë ¥ì´ í¬ê¸°
print("íšŒê·€ê³„ìˆ˜:", np.round(lr.coef_,1))
```
